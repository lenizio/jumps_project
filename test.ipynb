```python
from jumps.scm import Scm
from jumps.jump import Jump
from jumps.drop_jump import Drop_Jump
from jumps.lateral_jump import Lateral_Jump
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import xlsxwriter
from scipy import signal
```

```python
directoty = Path('/home/lenizio/biomechanics/jumps_project/data')
for file in directoty.rglob("*"):
    if file.is_file():
        name,salto,tentativa,peso = file.stem.strip().split('_')
        
        if salto=="scm":
            jump= Scm(name,salto,tentativa,peso,filepath=file)
            print(file)
            print('\n')
            # jump.plot_curve_ft()
            jump.plot_curves()
            jump.print_jump_data()
            
        if salto.startswith('sp'):
            drop_platform_height = salto.strip().split('.')[1]
            jump= Drop_Jump(name,salto,tentativa,peso,drop_platform_height,filepath=file)
            print(file)
            print('\n')
            # jump.plot_curve_ft()
            jump.plot_curves()
            jump.print_jump_data()
            
        if salto.startswith('sl'):
            hurdle_height = salto.strip().split('.')[1]
            jump=Lateral_Jump(name,salto,tentativa,peso,hurdle_height,filepath=file)
            jump.plot_curve_ft()
```
